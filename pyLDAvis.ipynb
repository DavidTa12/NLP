```python
from __future__ import print_function
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
import spacy
# To tag words
from textblob import TextBlob
import pandas as pd
from pandas import DataFrame
# To use new datatypes
from collections import Counter
from urllib import request
import re
import numpy as np
```

```python
url = "http://www.gutenberg.org/files/11/11-0.txt"
response = request.urlopen(url)
book = response.read().decode('utf8')
type(book)
# spacy.cli.download("en_core_web_lg")
# nlp = spacy.load('en_core_web_lg')
# #nlp = English()
```

```python
myChap = [
    book[1492:13267], book[13267:24430],
    book[24430:33906], book[33906:48059],
    book[48059:60374], book[60374:74550],
    book[74550:87604], book[87604:101587],
    book[101587:114545], book[114545:126262],
    book[126262:136916], book[136916:148851]]
```

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5, 
                                min_df = 0.1)
dtm_tf = tf_vectorizer.fit_transform(myChap)
print(dtm_tf.shape)
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(myChap)
print(dtm_tfidf.shape)
```

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_components=20, random_state=0)
lda_tf.fit(dtm_tf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=20, random_state=0)
lda_tfidf.fit(dtm_tfidf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer, mds='tsne')
```
